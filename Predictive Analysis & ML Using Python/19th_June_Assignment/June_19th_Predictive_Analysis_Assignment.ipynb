```python
# Impoting Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
# Other libraries
from sklearn.linear_model import LinearRegression 
```

```python
# Making the given data points into our desire dataset
x = np.array([2, 5, 1, 3, 4, 1, 5, 3, 4, 2])
y = np.array([50, 57, 41, 54, 54, 38, 63, 48, 59, 49])

df = pd.DataFrame({
    'commercials': x,
    'sales': y
})

df.to_csv('sales_data.csv', index=False)
```

```python
# Reading and loading our dataset
df = pd.read_csv('sales_data.csv')
```

```python
# Labelling our data features into X and y
X = df[['commercials']]
y = df['sales']
```

```python
# Visualizing our data 
plt.scatter(X, y, color='red', label='Sales Data')
plt.xlabel('Commercials')
plt.ylabel('Sales')
plt.title('Sales vs Commercials')
plt.legend()
plt.show()
```

```python
sns.lmplot(x='commercials', y='sales', data=df)
```

```python
# Calling the Model
model = LinearRegression()
```

```python
# Training the Model
model.fit(X, y)
```

```python
# Predicting the model
model.predict([[5]])
```

```python
# Regression line on the model
sns.scatterplot(x='commercials', y='sales', data=df)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('commercials')
plt.ylabel('sales')
plt.show()
```

```python
# Three things are important in SLR
# Slope, Intercept and r2 value
print('Intercept: ', model.intercept_)
print('Slope: ', model.coef_)
print('R2: ', model.score(X, y))
```

```python
import statsmodels.api as sm 
# Fit a linearregresssion model
model = sm.OLS(y, sm.add_constant(X)).fit()
# Get the P-value of the slope coefficient
p_value = model.params[1]
print('p_value: ', p_value)
```

```python
# !!!Evaluate Krty waqt different huga model!!! 
# Evaluation Metrics
# r2, mean squared error, mean absolute error, root mean squared error
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, train_size=0.8)

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
# Compare
sns.scatterplot(x=y_test, y=predictions)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE = mean_absolute_error(y_true=y_test, y_pred=predictions)
MSE = mean_squared_error(y_true=y_test, y_pred=predictions)
r2 = r2_score(y_true=y_test, y_pred=predictions)
# RMSE = mean_squared_error(y_true=y_test, y_pred=predictions, squared=False)

print('MAE:', MAE)
print('MSE:', MSE)
# print('RMSE:', RMSE)
print('r2:', r2)
```

```python
RMSE = np.sqrt(MSE)
print('RMSE:', RMSE)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22])
y = np.array([100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100])

df = pd.DataFrame({
    'Weekend Television Commercials': x,
    'Subsequent Store Sales': y

})

df.to_csv('Another_Store_Sales_Data.csv', index=False)
```

```python
df1 = pd.read_csv('Another_Store_Sales_Data.csv')
```

```python
X1 = df1[['Weekend Television Commercials']]
y1 = df1['Subsequent Store Sales']
```

```python
plt.title('Task 2: Commercials vs. Sales')
plt.scatter(X1, y1, color='Blue', label= 'Observed Data Points')
plt.xlabel('Weekend Television Commercials')
plt.ylabel('Subsequent Store Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
sns.lmplot(data=df1, x='Weekend Television Commercials', y='Subsequent Store Sales')
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()

model1.fit(X1, y1)

model1.predict(X1)
```

```python
# Regression line on the model
sns.scatterplot(x='Weekend Television Commercials', y='Subsequent Store Sales', data=df1)
plt.plot(X1, model1.predict(X1), color='red')
plt.xlabel('Weekend Television Commercials')
plt.ylabel('Subsequent Store Sales')
plt.show()
```

```python
# Three things are important in SLR
# Slope, Intercept and r2 value
print('Intercept: ', model1.intercept_)
print('Slope: ', model1.coef_)
print('R2: ', model1.score(X1, y1))
```

```python
import statsmodels.api as sm 
# Fit a linearregresssion model
model1 = sm.OLS(y1, sm.add_constant(X1)).fit()
# Get the P-value of the slope coefficient
p_value = model1.params[1]
print('p_value: ', p_value)
```

```python
# Evaluating the model 
# Evaluation Metrics
# r2, mean squared error, mean absolute error, root mean squared error
from sklearn.model_selection import train_test_split
X1_test, X1_train, y1_test, y1_train = train_test_split(X1, y1, train_size=0.8)

model2 = LinearRegression()
model2.fit(X1_train, y1_train)
predictions1 = model2.predict(X1_test)
```

```python
# Compare
sns.scatterplot(x=y1_test, y=predictions1)
```

```python
model2.score(X1_test, y1_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE = mean_absolute_error(y_true=y1_test, y_pred=predictions1)
MSE1 = mean_squared_error(y_true=y1_test, y_pred=predictions1)
r2 = r2_score(y_true=y1_test, y_pred=predictions1)
# RMSE = mean_squared_error(y_true=y_test, y_pred=predictions, squared=False)

print('MAE:', MAE)
print('MSE:', MSE)
# print('RMSE:', RMSE)
print('r2:', r2)
```

```python
RMSE1 = np.sqrt(MSE1)
print('RMSE:', RMSE1)
```
