### Task 1: An electronic store manager seeks to determine the relationship between the number oftelevision commercials broadcast on a weekend (x) and the store sales during the following week(y). The corresponding data are given by the following listsx=[2,5,1,3,4,1,5,3,4,2] and y=[50,57,41,54,54,38,63,48,59,49].
- Visualize the data set.

choose and compute a statistical measure that supports your argument best. Provide the 
necessary theoretical background.

```python
# Impoting the Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
# Other libraries
from sklearn.linear_model import LinearRegression 
```

```python
# Making the given data points into our desire dataset
x = np.array([2, 5, 1, 3, 4, 1, 5, 3, 4, 2])
y = np.array([50, 57, 41, 54, 54, 38, 63, 48, 59, 49])

df = pd.DataFrame({
    'commercials': x,
    'sales': y
})
# Saving the dataframe to a csv file named 'sales_date', and not write this datasets index. 
df.to_csv('sales_data.csv', index=False)
```

```python
# Reading and loading our dataset
df = pd.read_csv('sales_data.csv')
```

```python
# Labelling our data features into X and y
X = df[['commercials']]
y = df['sales']
```

```python
# Visualizing our data 
# Creating a scatter plot using the X and y data.
plt.scatter(X, y, color='red', label='Sales Data')
# Adding the label to the x-asix
plt.xlabel('Commercials')
# Adding the label to the y-axis
plt.ylabel('Sales')
# Adding a title here
plt.title('Sales vs Commercials')
# Displaying the legend 
plt.legend()
# Saving the plot 
plt.savefig('sales vs commercials.png', dpi=300)
# Showing our plot
plt.show()
```

```python
# Checking here, whether we should apply lineear regression algorithm or not, using seaborn
lmp = sns.lmplot(x='commercials', y='sales', data=df)
# Adding the title to the entire plot
lmp.fig.suptitle('lmplot')
# Saving the plot
lmp.fig.savefig('lmplot.png', dpi=300)
# Displaying the plot
plt.show()
```

```python
# Calling the Model
model = LinearRegression()
```

```python
# Training the Model
model.fit(X, y)
```

```python
# Predicting the model
model.predict([[5]])
```

```python
# Regression line on the model
# Plotting the original data points.
sns.scatterplot(x='commercials', y='sales', data=df)
# Adding the model's regression line in red.
plt.plot(X, model.predict(X), color='red')
# Labelling the axes and title.
plt.xlabel('commercials')
plt.ylabel('sales')
plt.title('Regression Line')
# Saving the plot
plt.savefig('Fitted regression line', dpi=300)
# Display the plot
plt.show()
```

```python
# Three things are important in SLR
# Slope, Intercept and r2 value
print('Intercept: ', model.intercept_)
print('Slope: ', model.coef_)
print('R2: ', model.score(X, y))
```

```python
# Importing the libraries again
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the numpy array for the commercials and sales figures
x = np.array([1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22])
y = np.array([100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100])
# Creating a Dataframe using pandas 
df = pd.DataFrame({
    'Weekend Television Commercials': x,
    'Subsequent Store Sales': y

})
# Saving the new Dataframe to a CSVfile without its index 
df.to_csv('Another_Store_Sales_Data.csv', index=False)
```

```python
# Loading the data from the CSV file
df1 = pd.read_csv('Another_Store_Sales_Data.csv')
```

```python
# Selecting the feature (independent variable) to X1
X1 = df1[['Weekend Television Commercials']]
# Selecting the target (dependent variable) to y1
y1 = df1['Subsequent Store Sales']
```

```python
# Setting the title
plt.title('Task 2: Commercials vs. Sales')
# Creating the scatter plot
plt.scatter(X1, y1, color='Blue', label= 'Observed Data Points')
# Labelling the x and y axis
plt.xlabel('Weekend Television Commercials')
plt.ylabel('Subsequent Store Sales')
# Displaying the plot's legend
plt.legend()
# Adding the grid for more readability
plt.grid(True)
# Saving our fig
plt.savefig('Task_2_Commercials_vs_Sales.png', dpi=300)
# Showing the plot
plt.show()
```

```python
# Importing the seaborn library
import seaborn as sns
# Creating a scatter plot with a fitted regression line
linear_model_plot = sns.lmplot(data=df1, x='Weekend Television Commercials', y='Subsequent Store Sales')
# Add a main title to the plot
linear_model_plot.fig.suptitle('LM plot of Another store')
# Saving the plot
linear_model_plot.fig.savefig('lm_plot_of_another.png', dpi=300)
# Showing our plot
plt.show()
```

```python
# Importing the LinearRegression model from scikit-learn 
from sklearn.linear_model import LinearRegression
# Creating the model
model1 = LinearRegression()
# Training the mode
model1.fit(X1, y1)
# Predicting the model
model1.predict(X1)
```

```python
# Three things are important in SLR
# Slope, Intercept and r2 value
# # Printing the model's intercept (line crossing the y-axis)
print('Intercept: ', model1.intercept_)
# Printing the model's slope (the steepness)
print('Slope: ', model1.coef_)
# Printing the R-squared value, (how the model fits the data)
print('R2: ', model1.score(X1, y1))
```

# Choosing the Polynomial Regression Model for My Another_Store_Sales_Data And Why?

---

```python
# Importing libraries again
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
# Importing other libraries 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Setting up our Data
# This data is from task 2
# Creating the dataframe using the numpy array
# Assign the new labels so overlapping don't happen
x_data = np.array([1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]).reshape(-1,1)
y_data = np.array([100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100])
```

```python
# Successfully created a 2-dimensional data
x_data
```

```python
# Visualizing rhe data, to get the glimpse of it
# Setting the main title for plot
plt.title('EDA Task 2: Commercials vs Sales')
# Creating the scatter plot of the observed data points
plt.scatter(x_data, y_data, color='Blue', label= 'Observed Data Points')
# Adding labels for the x and y-asix
plt.xlabel('Weekend Television Commercials')
plt.ylabel('Subsequent Store Sales')
# Displaying the legend
plt.legend()
# Adding the grid for more readability
plt.grid(True)
# Saving the plot
plt.savefig('EDA on task 2.png', dpi=300)
# Showing the plot
plt.show()
```

### As we can clearly see, there is this curve it appears to be dropping below and then again rising look like a U-Shaped.

```python
# Now, I'm going to create and train the model
# Creating the polynomial features to degree 2 
poly_features = PolynomialFeatures(degree=2, include_bias=False)
# Here, transforming the original feature data into polynomial features
x_poly = poly_features.fit_transform(x_data)

# Fitting a linear regression model to the polynomial features
# Creating a new regression model
model3 = LinearRegression()
# Training the model on polynomial features
model3.fit(x_poly, y_data)
```

```python
# Using the trained polynomial model to make predictions
y3_predict = model3.predict(x_poly)
```

```python
# Showing the result
y3_predict
```

```python
# To create a smooth curve, we generate many points between the min and max x values.
x_curve = np.linspace(x_data.min(), x_data.max(), 200).reshape(-1, 1)

# We must transform these new x-points into polynomial features as well.
x_curve_poly = poly_features.transform(x_curve)

# Predict the y-values for the curve using the trained model.
y_curve = model3.predict(x_curve_poly)
```

```python
plt.Figure(figsize=(10,8))
plt.scatter(x_data, y_data, color='red', label='Actual Sales Data')
# Plot the original data points as a scatter plot (blue dots).
plt.scatter(x_data, y_data, color='blue', label='Original Sales Data')

# Plot the calculated regression curve (red line).
plt.plot(x_curve, y_curve, color='red', linewidth=2, label='Polynomial Regression Line')

# Adding labels and a title to make the plot.
plt.title('Store Sales vs. TV Commercials')
plt.xlabel('Number of TV Commercials')
plt.ylabel('Store Sales')
# Saving the plot
plt.savefig('representing_the_polynomial_curve.png', dpi=300)
# Displaying the legend
plt.legend()
# Shwoing the grid for more readability
plt.grid(True)
# Dsiplaying the plot
plt.show()
```

```python
print('Intercept: ', model3.intercept_)
```

```python
# Now the evalaution of this Polynomial Regression model
# Importing the metrics

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2_poly  =  r2_score(y_data, y3_predict)
mae_poly = mean_absolute_error(y_data, y3_predict)
mse_poly = mean_squared_error(y_data, y3_predict)

# Printing the MAE, MSE, r2 values
print('MAE:', mae_poly)
print('MSE:', mse_poly)
print('r2:', r2_poly)
```

```python
RMSE_poly = np.sqrt(mse_poly)
print('RMSE:', RMSE_poly)
```
