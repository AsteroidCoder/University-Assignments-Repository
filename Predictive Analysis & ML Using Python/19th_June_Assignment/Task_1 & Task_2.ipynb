```python
# Impoting Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
# Other libraries
from sklearn.linear_model import LinearRegression 
```

```python
# Making the given data points into our desire dataset
x = np.array([2, 5, 1, 3, 4, 1, 5, 3, 4, 2])
y = np.array([50, 57, 41, 54, 54, 38, 63, 48, 59, 49])

df = pd.DataFrame({
    'commercials': x,
    'sales': y
})

df.to_csv('sales_data.csv', index=False)
```

```python
# Reading and loading our dataset
df = pd.read_csv('sales_data.csv')
```

```python
# Labelling our data features into X and y
X = df[['commercials']]
y = df['sales']
```

```python
# Visualizing our data 
plt.scatter(X, y, color='red', label='Sales Data')
plt.xlabel('Commercials')
plt.ylabel('Sales')
plt.title('Sales vs Commercials')
plt.legend()
plt.show()
```

```python
sns.lmplot(x='commercials', y='sales', data=df)
```

```python
# Calling the Model
model = LinearRegression()
```

```python
# Training the Model
model.fit(X, y)
```

```python
# Predicting the model
model.predict([[5]])
```

```python
# Regression line on the model
sns.scatterplot(x='commercials', y='sales', data=df)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('commercials')
plt.ylabel('sales')
plt.show()
```

```python
# Three things are important in SLR
# Slope, Intercept and r2 value
print('Intercept: ', model.intercept_)
print('Slope: ', model.coef_)
print('R2: ', model.score(X, y))
```

```python
import statsmodels.api as sm 
# Fit a linearregresssion model
model = sm.OLS(y, sm.add_constant(X)).fit()
# Get the P-value of the slope coefficient
p_value = model.params[1]
print('p_value: ', p_value)
```

```python
# !!!Evaluate Krty waqt different huga model!!! 
# Evaluation Metrics
# r2, mean squared error, mean absolute error, root mean squared error
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, train_size=0.8)

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
# Compare
sns.scatterplot(x=y_test, y=predictions)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE = mean_absolute_error(y_true=y_test, y_pred=predictions)
MSE = mean_squared_error(y_true=y_test, y_pred=predictions)
r2 = r2_score(y_true=y_test, y_pred=predictions)
# RMSE = mean_squared_error(y_true=y_test, y_pred=predictions, squared=False)

print('MAE:', MAE)
print('MSE:', MSE)
# print('RMSE:', RMSE)
print('r2:', r2)
```

```python
RMSE = np.sqrt(MSE)
print('RMSE:', RMSE)
```

# Choosing the Polynomial Regression Model for My Another_Store_Sales_Data And Why?

---

```python
# Beginning from the beginning 
# Importing libraries again

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Setting up our Data
# This data is from task 2

x_data = np.array([1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]).reshape(-1,1)
y_data = np.array([100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100])
```

```python
# As we can see, I have successfully create a 2-dimensional data
x_data
```

```python
# Visualizing rhe data, to get the glimpse of it

plt.title('EDA Task 2: Commercials vs Sales')
plt.scatter(x_data, y_data, color='Blue', label= 'Observed Data Points')
plt.xlabel('Weekend Television Commercials')
plt.ylabel('Subsequent Store Sales')
plt.legend()
plt.grid(True)
plt.show()
```

### As we can clearly see, there is this curve it appears to be dropping below and then again rising look like a U-Shaped.

```python
# Now, I'm going to create and train the model 
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_features.fit_transform(x_data)

# Fitting a linear regression model to the polynomial features
model3 = LinearRegression()
model3.fit(x_poly, y_data)
```

```python
# Predicting the value

y3_predict = model3.predict(x_poly)
```

```python
y3_predict
```

```python
# To create a smooth curve, we generate many points between the min and max x values.
x_curve = np.linspace(x_data.min(), x_data.max(), 200).reshape(-1, 1)

# We must transform these new x-points into polynomial features as well.
x_curve_poly = poly_features.transform(x_curve)

# Predict the y-values for the curve using the trained model.
y_curve = model3.predict(x_curve_poly)
```

```python
plt.Figure(figsize=(10,8))
plt.scatter(x_data, y_data, color='red', label='Actual Sales Data')
# Plot the original data points as a scatter plot (blue dots).
plt.scatter(x_data, y_data, color='blue', label='Original Sales Data')

# Plot the calculated regression curve (red line).
plt.plot(x_curve, y_curve, color='red', linewidth=2, label='Polynomial Regression Line')

# Add labels and a title to make the plot.
plt.title('Store Sales vs. TV Commercials')
plt.xlabel('Number of TV Commercials')
plt.ylabel('Store Sales')

plt.legend()
plt.grid(True)
plt.show()
```

```python
print('Intercept: ', model3.intercept_)
```

```python
# Now the evalaution of this Polynomial Regression model
# Importing the metrics

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2_poly  =  r2_score(y_data, y3_predict)
mae_poly = mean_absolute_error(y_data, y3_predict)
mse_poly = mean_squared_error(y_data, y3_predict)


print('MAE:', mae_poly)
print('MSE:', mse_poly)
print('r2:', r2_poly)
```

```python
RMSE_poly = np.sqrt(mse_poly)
print('RMSE:', RMSE_poly)
```
