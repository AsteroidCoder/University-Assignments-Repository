```python
# Impoting Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
# Other

from sklearn.linear_model import LinearRegression 
```

```python
x = np.array([2, 5, 1, 3, 4, 1, 5, 3, 4, 2])
y = np.array([50, 57, 41, 54, 54, 38, 63, 48, 59, 49])

df = pd.DataFrame({
    'commercials': x,
    'sales': y
})

df.to_csv('sales_data.csv', index=False)
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
X = df[['commercials']]
y = df['sales']
```

```python
plt.scatter(X, y, color='red', label='Sales Data')
plt.xlabel('Commercials')
plt.ylabel('Sales')
plt.title('Sales vs Commercials')
plt.legend()
plt.show()
```

```python
sns.lmplot(x='commercials', y='sales', data=df)
```

```python
# Calling the Model
model = LinearRegression()
```

```python
# Training the Model
model.fit(X, y)
```

```python
# Predicting the model
model.predict([[5]])
```

```python
# !!!Evaluate Krty waqt different huga model!!! 
# Evaluation Metrics
# r2, mean squared error, mean absolute error, root mean squared error
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(X, y, train_size=0.8)

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
# Compare
sns.scatterplot(x=y_test, y=predictions)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE = mean_absolute_error(y_true=y_test, y_pred=predictions)
MSE = mean_squared_error(y_true=y_test, y_pred=predictions)
r2 = r2_score(y_true=y_test, y_pred=predictions)
# RMSE = mean_squared_error(y_true=y_test, y_pred=predictions, squared=False)

print('MAE:', MAE)
print('MSE:', MSE)
# print('RMSE:', RMSE)
print('r2:', r2)
```

```python
RMSE = np.sqrt(MSE)
print('RMSE:', RMSE)
```
